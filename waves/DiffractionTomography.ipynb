## Forward Scatter Geometry

The figure below illustrates a forward scattering geometry. Here the illumnation field has no $x$ or $y$ wavevector components and $u_{z0}=\frac{1}{\lambda}$. The Fourier space in the $u_xu_z$ plane covered by the scatter is the curve $(u_x,\sqrt{\frac{1}{\lambda}-u_x^2})$ for $u_x\leq \frac{X}{\lambda z}$.

```

# arc observed with forward scatter
plt.rcParams["figure.autolayout"] = True
plt.rcParams['figure.figsize'] = [20, 20]
fig, ax = plt.subplots()
arc1 = patches.Arc((-uo,0),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
plt.gca().add_patch(arc1)
plt.ylim([-.2,.2])
plt.xlim([-.02,0])
plt.gca().set_aspect('equal')
xtick_loc = [-.02,0]
ax.set_xticks(xtick_loc)

```

```
# arc observed with forward scatter
plt.rcParams["figure.autolayout"] = True
plt.rcParams['figure.figsize'] = [40, 20]
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
arc1 = patches.Arc((-uo,0),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
ax1.add_patch(arc1)
ax1.set_ylim([-.2,.2])
ax1.set_xlim([-.02,0])
ax1.set_aspect('equal')
xtick_loc = [-.02,0]
ax1.set_xticks(xtick_loc)
ax1.set_title('(a)')
#  rotating object
ax2.set_ylim([-.2,.2])
ax2.set_xlim([-.1,.1])
for pip in range(-3,4):
  arc= patches.Arc((-uo,0),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
  t2 = mpl.transforms.Affine2D().rotate_deg(-pip*5)+ax2.transData
  arc.set_transform(t2)
  ax2.add_patch(arc)
ax2.set_title('(b)')
# scaling wavelength 
ax3.set_ylim([-.2,.2])
ax3.set_xlim([-.1,.1])
arcg= patches.Arc((-uo,0),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5, color='green')
ax3.add_patch(arcg)
arcb= patches.Arc((-.8*uo,0),2*.8*uo,2*.8*uo,angle=0,theta1=-10,theta2=10,linewidth=5, color='blue')
ax3.add_patch(arcb)
arcr= patches.Arc((-1.2*uo,0),2*1.2*uo,2*1.2*uo,angle=0,theta1=-10,theta2=10,linewidth=5, color='red')
ax3.add_patch(arcr)
ax3.set_title('(c)')
plt.savefig('forwardScatterShells.pdf')
files.download('forwardScatterShells.pdf')
```

## rotating the object with Forward scatter

If we rotate the object within the same sampling geometry, the sampled slice in Fourier Space rotates as well

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.ylim([-.2,.2])
plt.xlim([-.1,.05])
for pip in range(-3,4):
  arc= patches.Arc((-uo,0),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
  t2 = mpl.transforms.Affine2D().rotate_deg(-pip*5)+ax.transData
  arc.set_transform(t2)
  plt.gca().add_patch(arc)
plt.gca().set_aspect('equal')
```

## rotating the illumination beam angle with forward scatter

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.ylim([-1,1])
plt.xlim([-.5,.5])
for pip in range(-3,4):
  arc= patches.Arc((-uo*np.cos(pip/5),uo*np.sin(pip/5)),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
  plt.gca().add_patch(arc)
plt.gca().set_aspect('equal')
```

## rotating the illumination angle with back scatter

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.ylim([-1,1])
plt.xlim([1.5,2])
for pip in range(-3,4):
  arc= patches.Arc((uo*np.cos(pip/5),uo*np.sin(pip/5)),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
  plt.gca().add_patch(arc)
plt.gca().set_aspect('equal')
```

## scanning the illumination wavelength with back scatter

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.ylim([-.25,.25])
plt.xlim([1.5,2])
for pip in range(-10,0):
  us=uo*(1+pip/50.)
  arc= patches.Arc((us,0),2*us,2*us,angle=0,theta1=-10,theta2=10,linewidth=5)
  plt.gca().add_patch(arc)
plt.gca().set_aspect('equal')
```

```
# shifting illumination forward scatter
plt.rcParams["figure.autolayout"] = True
plt.rcParams['figure.figsize'] = [40, 20]
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
ax1.set_ylim([-1,1])
ax1.set_xlim([-.5,.5])
for pip in range(-3,4):
  arc= patches.Arc((-uo*np.cos(pip/5),uo*np.sin(pip/5)),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
  ax1.add_patch(arc)
ax1.set_title('(a)')
# shifting illumination angle back scatter
ax2.set_ylim([-1,1])
ax2.set_xlim([1.5,2])
for pip in range(-3,4):
  arc= patches.Arc((uo*np.cos(pip/5),uo*np.sin(pip/5)),2*uo,2*uo,angle=0,theta1=-10,theta2=10,linewidth=5)
  ax2.add_patch(arc)
ax2.set_title('(b)')
# scaling wavelength with back scatter
ax3.set_ylim([-.25,.25])
ax3.set_xlim([1.5,2])
for pip in range(-10,0):
  us=uo*(1+pip/50.)
  arc= patches.Arc((us,0),2*us,2*us,angle=0,theta1=-10,theta2=10,linewidth=5)
  ax3.add_patch(arc)
ax3.set_aspect('equal')
ax3.set_title('(c)')
plt.savefig('backScatterShells.pdf')
files.download('backScatterShells.pdf')
```

## bandpass and reconstruction example with backscatter

The Shepp Logan phantom is 400 by 400 pixels, assuming sampling at the rate $\frac{\lambda}{4}$ it is $100\times 100 \lambda $ in extent. We consider here measurement of the phantom using diffraction tomography in the $xz$ plane with wavelength scanning in the reflection geometry. The Fourier transform of the phantom spans the region $-\frac{2}{\lambda}, \frac{2}{\lambda}$ along the $u$ axis. However, the scatter covers the region only from $\pm u_{x\ \rm max}$ and from $\frac{1.6}{\lambda}$ to $\frac{2}{\lambda}$ along $u_z$. We assume that $u_{x\ \rm max}=\frac{.5}{\lambda}$

```
.2*512
```

```
from skimage.data import shepp_logan_phantom
```

```
plt.rcParams["figure.autolayout"] = True
plt.rcParams['figure.figsize'] = [20, 20]
image=np.zeros((1024,1024))
image[312:712,312:712] = shepp_logan_phantom()
image=image-np.mean(image)
imageFT=np.fft.fftshift(np.fft.fft2(image))
fig, ((ax1, ax2), (ax3, ax4) )=plt.subplots(2,2)
ax1.imshow(image[312:712,312:712], extent=[-50,50,-50,50], cmap='gray')
ax1.set_xlabel(r'$\lambda$')
#
#
ax2.imshow(np.abs(imageFT)[312:712,312:712], extent=[-2*100/512,2*100/512,-2*100/512,2*100/512], cmap='gray')
ax2.set_xlabel(r'$\frac{1}{\lambda}$')
measuredData=imageFT[(512-128):(512+128),922:1024]
ax3.imshow(np.abs(measuredData), extent=[2*412/512,2,-.5,.5], cmap='gray')
ax3.set_xlabel(r'$\frac{1}{\lambda}$')
imFT=np.zeros((1024,1024)).astype('complex128')
imFT[(512-128):(512+128),(512-51):(512+51)]=measuredData
ax4.imshow(np.abs(np.fft.ifft2(imFT))[312:712,312:712], extent=[-50,50,-50,50], cmap='gray')
ax4.set_xlabel(r'$\lambda$')
plt.savefig('backScatterReconstruction.pdf')
files.download('backScatterReconstruction.pdf')
```
